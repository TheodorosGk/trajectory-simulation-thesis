```python
import pandas as pd
from sklearn.model_selection import train_test_split

df = pd.read_csv(
    r'C:\Users\thodo\Desktop\DIPLOMATIKI ERGASIA\EFARMOGI\TS_TRAJGEN DATASET DONLOADED FROM USER\TS-TrajGen_Dataset\Porto_Taxi\TS_TrajGen_data_archive\pretrain_data\Porto_Taxi_trajectory.csv'
)
```

```python
df.head()
```

```python

#### GIA EKSIKONOMISI XRONOU THA PAROUME TO 15% TOU ORIGINAL DATA
traj_ids = df["traj_id"].unique()

# ΠΑΡΕ ΜΟΝΟ ΤΟ 15%
sample_ids, _ = train_test_split(
    traj_ids,
    test_size=0.85,
    random_state=42,
    shuffle=True
)

df_10 = df[df["traj_id"].isin(sample_ids)].copy()

print(f"Original trajectories: {len(traj_ids)}")
print(f"15% trajectories: {df_10['traj_id'].nunique()}")
```

```python
traj_ids_10 = df_10["traj_id"].unique()

train_ids, test_ids = train_test_split(
    traj_ids_10,
    test_size=0.2,
    random_state=42,
    shuffle=True
)

train_df = df_10[df_10["traj_id"].isin(train_ids)].copy()
test_df  = df_10[df_10["traj_id"].isin(test_ids)].copy()
```

```python
import os
out_dir = r'C:\Users\thodo\Desktop\DIPLOMATIKI ERGASIA\EFARMOGI\TS_TRAJGEN DATASET DONLOADED FROM USER\TS-TrajGen_Dataset\Porto_Taxi'

train_df.to_csv(
    os.path.join(out_dir, 'porto_mm_train.csv'),
    index=False
)

test_df.to_csv(
    os.path.join(out_dir, 'porto_mm_test.csv'),
    index=False
)
```

```python
# Καμία τροχιά να μην εμφανίζεται και στα δύο
assert set(train_df.traj_id).isdisjoint(set(test_df.traj_id))
```

```python
# Μήκη τροχιών
train_df["rid_len"] = train_df["rid_list"].apply(lambda x: len(x.split(',')))
test_df["rid_len"]  = test_df["rid_list"].apply(lambda x: len(x.split(',')))

print(train_df["rid_len"].describe())
print(test_df["rid_len"].describe())
```

```python
train_trajs = set(train_df["traj_id"])
test_trajs  = set(test_df["traj_id"])
```

```python
common = train_trajs.intersection(test_trajs)
print("Κοινές τροχιές:", len(common))
```
